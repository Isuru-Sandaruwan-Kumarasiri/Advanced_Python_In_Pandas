# Data Cleaning & Manipulation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

1.Load the dataset and convert the Formatted Date column into a proper datetime format.

```python
dataset = pd.read_csv('data.csv')
dataset.head()

```

2.Fill missing values in the Precip Type column with the most frequent value.

```python
missing_values_count = dataset['Precip Type'].isna().sum()
print("missing values count :",missing_values_count)

most_frequent_value = dataset['Precip Type'].mode()[0]
print("most_frequent_value :",most_frequent_value)

dataset['Precip Type'].fillna(most_frequent_value, inplace=True)
missing_values_count = dataset['Precip Type'].isna().sum()
print("missing values count :",missing_values_count)
```

3.Create a new column Temperature (F) that converts Temperature (C) to Fahrenheit.

```python
dataset['Temperature (F)'] = dataset['Temperature (C)'] * 9/5 + 32
dataset.head()
```

4.Extract the year and month from the Formatted Date column and create separate columns for them.

```python
# Key Points:
# Purpose

# Pandas provides an option (utc=True or utc=False) to control how time zone-related datetime values are handled during conversion.

             # If utc=True (Recommended for Future Versions)

# The function always returns a timezone-aware timestamp in UTC.
# If the input datetime has no time zone (naive), it is assumed to be in UTC.
# If the input datetime has a time zone, it is converted to UTC.


             # If utc=False (Default Behavior)

# Time zone handling remains unchanged:
# Naive datetime (no timezone) stays naive.
# Timezone-aware datetime keeps its original offset (e.g., +05:30, -08:00).
# However, issues can arise with mixed time zones (e.g., due to Daylight Savings Time (DST)), leading to inconsistencies.

dataset['Formatted Date'] = pd.to_datetime(dataset['Formatted Date'], utc=True)  
dataset['Year'] = dataset['Formatted Date'].dt.year
dataset['Month'] = dataset['Formatted Date'].dt.month
dataset.head()
```

5.Group the data by year and find the average temperature for each year.

```python
dataset.groupby("Year")[["Temperature (C)"]].mean()
```
